**Calculando a correlação entre as váriaveis**

````python
data_base.loc[:, 'g0' : 'g771'].corr()
````

**Quando a correlação está próxima de (1, -1) significa que elas estão fortemente correlacionadas. O valor 1 significa que a correlação é diretamente proporcional, já o valor -1 indica que elas são inversamente proporcionais.**

**Quando essa correlação se aproxima de 0 elas não estão correlacionadas.**

**Criando uma Matriz de Correlações**

````python
# Compute the correlation matrix
corr_gen = data_base.loc[:, 'g0' : 'g50'].corr()
````

````python
import numpy as np

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_gen, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_gen, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

**Calculando a correlação entre as variaveis "c"**

````python
corr_celular = data_base.loc[:, 'c0' : 'c50'].corr()
````

````python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

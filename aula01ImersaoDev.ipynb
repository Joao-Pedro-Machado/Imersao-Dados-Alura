```python
import pandas as pd

url_data = 'https://github.com/alura-cursos/imersaodados3/raw/main/dados/dados_experimentos.zip'

data_base = pd.read_csv(url_data, compression='zip')
data_base
```

Desafio 06: Renomear as colunas tirando o hífen.

```python
data_base.columns = data_base.columns.str.replace("-", "")
data_base.head()
```

Desafio 07 Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt

frame_analyze = data_base['tratamento'].value_counts(normalize = True)*100

graphic = frame_analyze.plot.barh()
plt.xlabel('Porcentagem(%)')
plt.ylabel('Tipo de Tratamento')
plt.title('proporção de tratamento')

graphic.yaxis_inverted()
```

Desafio 07 Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt

frame_analyze = data_base['tratamento'].value_counts(normalize = True)*100

graphic = frame_analyze.plot.barh()
plt.xlabel('Porcentagem(%)')
plt.ylabel('Tipo de Tratamento')
plt.title('proporção de tratamento')

graphic.yaxis_inverted()
```

```python
map = {'droga' : 'composto'}
data_base.rename(columns=map, inplace=True)
data_base
```

```python
cod_compost = data_base['composto'].value_counts().index[0:5]
cod_compost
```

```python
data_base.query('composto in @cod_compost')
```

```python
current_data = data_base.query('composto in @cod_compost')
current_data
```

```python
current_data = data_base.query('composto in @cod_compost')
current_data.sort_values(by='composto')
current_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

current_data = data_base.query('composto in @cod_compost')
current_data.sort_values(by='composto')

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.countplot(x = 'composto', data=current_data)
ax.set_title('TOP 5 COMPOSTOS')
plt.show
```

**Desafio 02: Melhorar a visualização alterando tamanho da fonte...**

```python
sns.set(font_scale=1.3)
plt.figure(figsize=(10, 8))
ax = sns.countplot(x = 'composto', order=cod_compost, data=current_data)
ax.set_title('TOP 5 COMPOSTOS')
plt.show
```

**Desafio 04: Estudar sobre as estatíticas retornadas no .describe()**

```python
data_base['g0'].describe()
```

**Count:**  Quantidade de valores não nulos contados

**Mean:**   Médio dos valores contados

**std:**  Desvio padrão, valor que indica o quantos os valores contados se paroximam da média, quanto menor mais os valores se aproximam

**min:**  Valor mínimo encontrado

**max:** Valor máximo encontrado

**porcentagens:**  indicam a porcentagem das médias do valores presente no data frame

Descrição das colunas " g " do data_base

```python
data_base.loc[:, 'g0' : 'g771'].describe()
```

Transpondo a base de dados

```python
data_base.loc[:, 'g0' : 'g771'].describe().T
```

Tranformando a base de dados transposta em um histograma, no caso, utilizando sómente a coluna mean("média")

```python
data_base.loc[:, 'g0' : 'g771'].describe().T['mean'].hist(bins=30)
```

Alterando a estatística analisada

```python
data_base.loc[:, 'g0' : 'g771'].describe().T['min'].hist(bins=30)
```

```python
data_base.loc[:, 'g0' : 'g771'].describe().T['max'].hist(bins=30)
```

Alterando os dados análisados

```python
data_base.loc[:,'c0':'c99'].describe().T['mean'].hist(bins=50)
```
